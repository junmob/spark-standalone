# The 5 levels of Apache Spark remastered

### LEVEL 1: BASICS  
- 3 Different APIs:  
    - RDDs  
    - Dataframes
    - Datasets  
- Transformations:  
    - Map
    - Filter
    - GroupBy
    - Join  
- Actions:  
    - Count
    - Show
    - Collect
    - Insert into  
- Lazy Execution (Spark won't run any transformation, until an action is taken.)  

### LEVEL 2: OPTIMIZING SPARK QUERIES  
- Understand Shuffle
- Wide vs Narrow Dependency
- Cache() vs Persist() and Immutability of DataFrame
- Coalesce vs Repartition
- Understand Catalyst Optimizer  
- Master Columnar Storage  
- Parquet vs JSON vs CSV  
- Avoid UDFs unless absolutely necessary  

### LEVEL 3: MASTER SPARK PERFORMANCE TUNING  
- Learn about JOINS (Broadcast join is amazing if one small and one big dataset)  
- Master spark.sql.autoBroadcastJoinThreshold  
- Understand how Task Parallelism works (spark.sql.shuffle.partitions)  
- Skewed Data use Adaptive Execution  
- Understand Explain Plans  

### LEVEL 4: Cluster Resource Management  
- YARN vs Kubernetes, Standalone, EMR, Databricks, Serverlist  
- Dynamic Allocation  
- RDDs vs Dataframes (spoiler: always use Dataframes)
